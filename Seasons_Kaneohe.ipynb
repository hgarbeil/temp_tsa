```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('/Users/hg/Downloads/Temps_Kaneohe.csv',index_col=False)
#print(data.head())
tdata=data[data['PARAMETER']=='T2M']
types=data['PARAMETER'].unique()
print(types)
relhum=data[data['PARAMETER']=='RH2M']
wspeed=data[data['PARAMETER']=='WS10M']
wdir=data[data['PARAMETER']=='WD10M']
precip=data[data['PARAMETER']=='PRECTOTCORR']

         
```

```python
print(tdata.head())
tdata['summer']=tdata[['JUL','AUG','SEP']].mean(axis=1)
tdata['fall']=tdata[['OCT','NOV','DEC']].mean(axis=1)
tdata['winter']=tdata[['JAN','FEB','MAR']].mean(axis=1)
tdata['spring']=tdata[['APR','MAY','JUN']].mean(axis=1)
```

```python
print(tdata.head())
```

```python
plt.figure(figsize=[10,6])
plt.title('Kaneohe Temps')
plt.xlabel('Year')
plt.ylabel('Temp C')
plt.plot('YEAR','ANN',data=tdata)
plt.plot('YEAR','winter',data=tdata)
plt.plot('YEAR','fall',data=tdata)
plt.plot('YEAR','spring',data=tdata)
plt.plot('YEAR','summer',data=tdata)
plt.legend()
```
